### Bayeisan probabilities

- frequentist view

view probabilities in terms of the frequencies of random, repeatable events

- bayesian view

probabilities provide a quantification of uncertainty

sometime we dont have a chance to repeatedly obatin observations, in such cirsumstances, we would like to quantify our expression of uncertainty and make precise revisions of uncertainty in the light of new evidence, as well as subsequently to be able to take optimal actions or decisions as a consequence

**Bayes' theorem was used to convert a prior into a posterior by incorporating the evidence provided by the observed data**

$posterior \propto likelihood \times prior$

i.e. **prior** of a set of weights to curve fitting $p(\textbf{w})$

**likelihood**: $p(\mathcal{D}|\textbf{w})$ expresses how probable the observed data set is for different settings of the parameter vector $\textbf{w}$

**posterior**: $p(\textbf{w}|\mathcal{D})$ after observing the data $\mathcal{D}$, what is the probability of the weights, where

$p(\textbf{w}|\mathcal{D})=\frac{p(\mathcal{D}|\textbf{w})p(\textbf{w})}{p(\mathcal{D})}$

$p(\mathcal{D})=\int p(\mathcal{D}|\textbf{w})p(\textbf{w}) d\textbf{w}$

**in frequentist view**:

$\textbf{w}$ is considered to be a fixed parameter, whose value is determined by some form of 'estimator', and error bars on this estimate are obtained by considering the distribution of possible data sets $\mathcal{D}$

**maximum likelihood** is a **frequentist estimator** in which $\textbf{w}$ is set to the value that maximizes the likelihood function $p(\mathcal{D}|\textbf{w})$ (namely choosing the value of $\textbf{w}$ for which the probability of the observed data set is maximized)

**from Bayesian view**:

there is only a single data set $\mathcal{D}$, (namely the one that is actually observed), and the uncertainty in the parameters is expressed through a probability distribution over $\textbf{w}$


**drawbacks** of Bayesian view:

- performance heavily rely on the choice of prior

- Bayesian procedure requires marginalizing over the whole of parameter space

Bayesian tools:

- the development of sampling methods such as MCMC (flexible but small scale)

- dramatic improvements in the speed and memory capacity of computers 

- variational Bayes

- expectation propagation

**drawbacks** of **maximum likelihood**:

- the maximum likelihood approach systematically underestimates the variance of the distribution

**note**: **any of the quantities appearing in Baye's Theorem can be obtained from the joint distribution $p(\textbf{x}, \textbf{t})$ by either marginalization or conditioning w.r.t the appropriate variables** 

**some notes for curve fitting**:

before observing the data a **prior** of the parameters: $p(\textbf{w})$

after observing the data $\mathcal{D}=(t_1,...,t_N)$, we have a **likelihood**: $p(\mathcal{D}|\textbf{w})$

the **posterior**: $p(\textbf{w}|\mathcal{D})=\frac{p(\mathcal{D}|\textbf{w})p(\textbf{w})}{p(\mathcal{D})}$

### Decision Theory

Decision Theory tells us how to make optimal decisions given the appropriate probabilities

Given $\textbf{x}$ and $\textbf{t}$, we want to predict $t$ with new $x$

for regression, $\textbf{t}$ is continuous variables

for classification, $\textbf{t}$ is class labels

Three ways to solve decision problems:

1. **Generative models**: approaches that explicitly or implicitly model the distribution of inputs and outputs

2. **Discriminative models**: approaches that model the posterior probabilities directly 

model $p(C_k|\textbf{x})$ directly, then use decision theory to determine class membership for each new input $\textbf{x}$

3. **use Discriminant function $f(x)$**: $f(x)$ directly maps input onto a class label

in this case probabilities play no role

### Maximum likelihood Estimation

https://www.analyticsvidhya.com/blog/2021/09/maximum-likelihood-estimation-a-comprehensive-guide/

**Def**:a process of using data to find estimators for different parameters characterizing a distribution

**common estimators**:

- sample mean

- sample variance

but they are not enough

**basics for statistical modeling**:

$(E, \{ \mathbb{P}_{\theta} \}_{\theta \in \Theta})$

$E$ - sample space of an experiment

$\mathbb{P}_{\theta}$ - the family of probability-measure on $E$

$\theta$ - the set of unknown parameters that characterize the distribution $\mathbb{P}_{\theta}$

$\Theta$ - the parameter space

examples of **Bernoulli Distribution**:

$(\{0,1\}, \{Ber(p)\}_{p \in [0,1]})$, where $p$ is the parameter for mean or probability of success

**Exponential Distribution**:

$([0,\infty), \{Exp(\lambda)\}_{\lambda \in (0,\infty)} )$

**Identifiability**:

different values of a parameter must produce different probability distributions

i.e. for two parameters $\theta$ and $\theta'$, there must exist two different distributions $\mathbb{P}_{\theta}$ and $\mathbb{P}_{\theta'}$

$\theta \neq \theta' => \mathbb{P}_{\theta} \neq \mathbb{P}_{\theta'}$, $\mathbb{P}_{\theta}=\mathbb{P}_{\theta'} => \theta=\theta'$

**Total Variation (TV) Distance**: measures the worst deviation between two probability distributions for any subset of the sample space $E$:

$TV(\mathbb{P},\mathbb{Q})=\max_{A\subseteq E}|\mathbb{P}(A)-\mathbb{Q}(A)|$

in discrete case:

$TV(\mathbb{P},\mathbb{Q})=\max_{A\subseteq E}|\mathbb{P}(A)-\mathbb{Q}(A)|=\frac{1}{2} \sum_{x \in E} |p(x)-q(x)|$

in continuous case:

$TV(\mathbb{P},\mathbb{Q})=\max_{A\subseteq E}|\mathbb{P}(A)-\mathbb{Q}(A)|=\frac{1}{2} \int_{x \in E} |p(x)-q(x)| dx$

**properties of TV distance**:

- symmetry: $TV(\mathbb{P}, \mathbb{Q})=TV(\mathbb{Q}, \mathbb{P})$

- definiteness: $TV(\mathbb{P}, \mathbb{Q})=0 => \mathbb{P}=\mathbb{Q}$

- range: $0\leq TV(\mathbb{P},\mathbb{Q}) \leq 1$ as $0\leq \max_{A \subseteq E} |\mathbb{P}(A)-\mathbb{Q}(A)| \leq 1$

- triangle inequality: $TV(\mathbb{P}, \mathbb{Q}) \leq TV(\mathbb{P}, \mathbb{R})+TV(\mathbb{R}, \mathbb{Q})$

**example of estimator using TV distance**:

$\mathbb{P}_{\theta}$ - $\theta$ parameter we want to have

$\mathbb{P}_{\theta^*}$ - $\theta^*$ true value of the parameter $\theta$

by the **definiteness** and **identifiability**:

$TV(\mathbb{P}_{\theta},\mathbb{P}_{\theta^*})=0 => \mathbb{P}_{\theta}=\mathbb{P}_{\theta^*} => \theta=\theta^* $

hence

$\theta^* = \arg \min_{\theta \in \Theta} TV(\mathbb{P}_{\theta},\mathbb{P}_{\theta^*})$

However it's impossible to compute $TV(\mathbb{P}_{\theta},\mathbb{P}_{\theta^*})$ since we dont know $\theta^*$

instead we can estimate the TV distance $\hat{TV}(\mathbb{P}_{\theta},\mathbb{P}_{\theta^*})$

then we can have a $\hat{\theta}$ which is close to $\theta^*$:

$\hat{\theta}=\arg \min_{\theta \in \Theta} \hat{TV}(\mathbb{P}_{\theta},\mathbb{P}_{\theta^*})$

but it's also difficult to find this $\hat{TV}$

**Kullback-Leibler Divergence**:

discrete: $KL(\mathbb{P} || \mathbb{Q})=\sum_{x \in E} p(x) \log \frac{p(x)}{q(x)}$

continuous: $KL(\mathbb{P} || \mathbb{Q})=\int_{x \in E} p(x) \log \frac{p(x)}{q(x)} dx$

**properties of KL divergence**:

- asymmetry: $KL(\mathbb{P} || \mathbb{Q}) \neq KL(\mathbb{Q} || \mathbb{P})$

- definiteness: $KL(\mathbb{P} || \mathbb{Q})=0 => \mathbb{P}=\mathbb{Q}$

- range: $KL(\mathbb{P} || \mathbb{Q}) \geq 0$

- no triangle inequality: $KL(\mathbb{P} || \mathbb{Q}) \nleq KL(\mathbb{P} || \mathbb{R})+KL(\mathbb{R} || \mathbb{Q})$

- amenable to estimation: $KL(\mathbb{P} || \mathbb{Q}) = \int_{x \in E} p(x) \log \frac{p(x)}{q(x)} dx = \mathbb{E}_{x \sim \mathbb{P}} \left[\log \frac{p(x)}{q(x)} \right]$

now to construct an estimator:

$KL(\mathbb{P}_{\theta^*}||\mathbb{P}_{\theta})=0 => \mathbb{P}_{\theta^*}=\mathbb{P}_{\theta} => \theta^*=\theta$

$\theta^* = \arg \min_{\theta \in \Theta} KL(\mathbb{P}_{\theta^*}||\mathbb{P}_{\theta})$

again we use an estimator $\hat{KL}$ to obtain $\hat{\theta}$:

$\hat{\theta} = \arg \min_{\theta \in \Theta} \hat{KL}(\mathbb{P}_{\theta^*}||\mathbb{P}_{\theta})$

**To construct a MLE estimator**:

starting from $KL(\mathbb{P}||\mathbb{Q})=E_{x \sim \mathbb{P}} \left[\log \frac{p(x)}{q(x)} \right]$

$
\begin{align}
KL(\mathbb{P}_{\theta^*}||\mathbb{P}_{\theta}) &= E_{x \sim \mathbb{P}_{\theta^*}} \left[\log \frac{\mathbb{P}_{\theta^*}(x) }{\mathbb{P}_{\theta}(x)} \right] \\
&= E_{x \sim \mathbb{P}_{\theta^*}} \left[\log \mathbb{P}_{\theta^*}(x) - \log \mathbb{P}_{\theta}(x) \right] \\
&= E_{x \sim \mathbb{P}_{\theta^*}} \left[\log \mathbb{P}_{\theta^*}(x) \right] - E_{x \sim \mathbb{P}_{\theta^*}} \left[ \log \mathbb{P}_{\theta}(x) \right] \\
&= c-E_{x \sim \mathbb{P}_{\theta^*}} \left[ \log \mathbb{P}_{\theta}(x) \right]
\end{align}
$

$c=E_{x \sim \mathbb{P}_{\theta^*}} \left[\log \mathbb{P}_{\theta^*}(x) \right]$ is treated as constant since it is independent of $\theta$

our goal is to minimize the KL divergence over $\theta$, and now we have

$\hat{KL}(\mathbb{P}_{\theta^*}||\mathbb{P}_{\theta})=c-\hat{E}_{x \sim \mathbb{P}_{\theta^*}} \left[ \log \mathbb{P}_{\theta}(x) \right]$

Recall the **law of large numbers**: 

as our sample size grows bigger, the sample mean of the observations converges to the true mean or expectation of the underlying distribution:

$\frac{1}{n} \sum_{i=1}^n Y_i \underset{n \rightarrow \infty}{\rightarrow} E(Y)$

we can replace $Y_i$ with $\log p_{\theta}(X_i)$

$\frac{1}{n} \sum_{i=1}^n \log p_{\theta}(X_i) \underset{n \rightarrow \infty}{\rightarrow} E_{x \sim \mathbb{p}_{\theta^*}}(\log p_{\theta}(X_i))$

we can use $\frac{1}{n} \sum_{i=1}^n \log p_{\theta}(x_i)$ as an estimator of $\hat{E}_{x \sim \mathbb{P}_{\theta^*}} \left[ \log \mathbb{P}_{\theta}(x) \right]$

so now

$\hat{KL}(\mathbb{P}_{\theta^*}||\mathbb{P}_{\theta})=c-\frac{1}{n} \sum_{i=1}^n \log p_{\theta}(x_i)$

$
\begin{align}
\hat{\theta} &= \arg \min_{\theta \in \Theta} \hat{KL}(\mathbb{P}_{\theta^*}||\mathbb{P}_{\theta}) \\
&= \arg \min_{\theta \in \Theta} -\frac{1}{n} \sum_{i=1}^n \log p_{\theta}(x_i) \\
&= \arg \max_{\theta \in \Theta} \frac{1}{n} \sum_{i=1}^n \log p_{\theta}(x_i) \\
&= \arg \max_{\theta \in \Theta} \sum_{i=1}^n \log p_{\theta}(x_i) \\
&= \arg \max_{\theta \in \Theta} \log \prod_{i=1}^n p_{\theta}(x_i) \\
&= \arg \max_{\theta \in \Theta} \prod_{i=1}^n p_{\theta}(x_i)
\end{align}
$

since $\log(x)$ is an increasing function

now we have our MLE estimator of the parameter $\theta$ as

$\hat{\theta}^{MLE} = \arg \max_{\theta \in \Theta} \prod_{i=1}^n p_{\theta}(x_i)$

where $\prod_{i=1}^n p_{\theta}(x_i)$ is called **likelihood function**

### MLE from Bayesian view

MLE also can be seen as a special case of **max a posteriori estimation (MAP)** that assumes a **uniform prior distribution** of the parameters, or as a variant of the MAP that ignores the prior and which therefore is **unregularized**

suppose 

$p(\theta|x)=\frac{p(x|\theta)p(\theta)}{p(x)}$

posterior=likelihood x prior / evidence

we can ignore the prior and evidence since they are assumed to be uniform

so **max the posterior $p(\theta|x)$ <=> max the likelihood $p(x|\theta)$**

**merits of MLE**:

- MLE is generalizable for regression and classification

- MLE is efficient (no consistent estimator has lower asymptotic error than MLE if you are using the right distribution)

- MLE can be regarded as a modular way of fitting models by optimizing a probabilistic cost function

**steps of MLE**:

1. define the likelihood by choosing the right distribution for your regression or classification problem

2. take the log and reduce the product function into a sum function 

3. max or min the negative of the logsum likelihood/objective function 

4. verify that uniform priors are a safe assumption

otherwise fails the Law of Parsimony

### Bernoulli distribution

binary random variable $x \in \{0, 1\}$

$p(x=1|\mu)=\mu$

$p(x=0|\mu)=1-\mu$

where $0 \leq \mu \leq 1$

$Ber(x|\mu)=\mu^x(1-\mu)^{1-x}$

when $x=0$, $Ber(x=0|\mu)=1-\mu$

when $x=1$, $Ber(x=1|\mu)=\mu$

$\mathbb{E}[x]=\sum_{0,1} Ber(x)x = Ber(0)0+Ber(1)1= \mu$

$var[x]=\mathbb{E}[x^2]-\mathbb{E}^2[x]=[\sum_{0,1} Ber(x)x^2]-\mu^2=Ber(1)-\mu^2=\mu(1-\mu)$

suppose we have an i.i.d data set $\mathcal{D}_{1:N}$ of observed values $x$

we can construct the likelihood function

$p(\mathcal{D}|\mu)=\prod_{n=1}^N p(x_n|\mu)=\prod_{n=1}^N \mu^{x_n}(1-\mu)^{1-x_n}$

by maximizing the log of this likelihood function

$\log p(\mathcal{D}|\mu)=\sum_{n=1}^N \log p(x_n|\mu)=\sum_{n=1}^N [x_n \log\mu+(1-x_n)\log(1-\mu)]$

and making its derivative w.r.t $\mu$ to 0

we can obtain a MLE estimator, also named **sample mean**

$\mu^{MLE}=\frac{1}{N}\sum_{n=1}^N x_n=\frac{m}{N}$

where $m$ is the number of observations of $x=1$

**problem of MLE**:

If we have three times of heads, the MLE estimator would tell us all future observations will be heads. This is an extreme example of the **over-fitting** associated with MLE

We can relieve this issue under a Bayesian view

### Binomial distribution

we can extend the binary variable $x\in \{0,1\}$ to a set of discrete numbers $m\in \{0,1,2,3,4,...\}$

Binomial distribution is the distribution of number $m$ of observatins of $x=1$, namely, the number of appearance of number $m$ in $N$ trials

$Bin(m|N,\mu)= \begin{pmatrix} N \\ m \end{pmatrix}\mu^m (1-\mu)^{N-m}$

$\begin{pmatrix} N \\ m \end{pmatrix}=C_N^m=\frac{N!}{(N-m)!m!}$ - the number of ways of choosing $m$ objects out of a total of $N$ identical objects

$\mathbb{E}[m]=\sum_{m=0}^N m Bin(m|N,\mu)=N\mu$

$var[m]=\sum_{m=0}^N(m-\mathbb{E}[m])^2Bin(m|N,\mu)=N\mu(1-\mu)$

**example of Binomial distribution**:

$m$ - number of heads from flipping a fair coin 5 times

    1st: H,T,T,H,H <-m=3
    2nd: H,T,T,T,H <-m=2
    3rd: T,T,T,H,H <-m=2
    4th: H,T,T,T,T <-m=1
    5th: H,T,H,H,H <-m=4
    
possible outcomes from 5 flips: 2^5=32

$p(m=0)=\frac{1}{32}=\frac{C_5^0}{32}$

    T,T,T,T,T 
    
$p(m=1)=\frac{5}{32}=\frac{C_5^1}{32}$

    H,T,T,T,T
    T,H,T,T,T
    T,T,H,T,T
    T,T,T,H,T
    T,T,T,T,H
    
$p(m=2)=\frac{10}{32}=\frac{C_5^2}{32}$

$p(m=3)=\frac{10}{32}=\frac{C_5^3}{32}$

$p(m=4)=\frac{5}{32}=\frac{C_5^4}{32}$

$p(m=5)=\frac{1}{32}=\frac{C_5^5}{32}$

back to our Binomial distribution

$Bin(m=0|N=5,\mu=0.5)=C_5^0 0.5^0 0.5^5=1/32$

$Bin(m=1|N=5,\mu=0.5)=C_5^1 0.5^1 0.5^4=5/32$

$Bin(m=2|N=5,\mu=0.5)=C_5^2 0.5^2 0.5^3=10/32$

$Bin(m=3|N=5,\mu=0.5)=C_5^3 0.5^3 0.5^2=10/32$

$Bin(m=4|N=5,\mu=0.5)=C_5^4 0.5^4 0.5^1=5/32$

$Bin(m=5|N=5,\mu=0.5)=C_5^5 0.5^5 0.5^0=1/32$

```python
def fact(n):
    return np.math.factorial(n)

def binomial(n,m,mu):
    return (fact(n)/(fact(n-m)*fact(m)))*(mu**m)*(1-mu)**(n-m)

def draw_binomial(n=10,mu=0.25):

    dist=[]
    for m in range(n+1):
        dist.append(binomial(n,m,mu))

    plt.bar(range(n+1),dist)
    
draw_binomial(n=10,mu=0.25)
```

### Beta Distribution

$Beta(\mu|a,b)=\frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)}\mu^{a-1}(1-\mu)^{b-1}$

$\frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)}$ ensures Beta distribution is normalized

$\mathbb{E}[\mu]=\frac{a}{a+b}$

$var[\mu]=\frac{ab}{(a+b)^2(a+b+1)}$

$\Gamma(x)=\int_0^{\infty} u^{x-1}e^{-u}du$ - gamma function

properties:

- $\Gamma(x+1)=x\Gamma(x)$

- $\Gamma(1)=1$ hence $\Gamma(x+1)=x!$ when $x$ is an integer

back to the **problem of MLE** in Bernoulli distribution:

$\mu^{MLE}=\frac{m}{N}$

If we have three times of heads, the MLE estimator would tell us all future observations will be heads. This is an extreme example of the **over-fitting** associated with MLE

We can relieve this issue under a **Bayesian view**

recall the likelihood function of Bernoulli:

$p(\mathcal{D}|\mu)=\prod_{n=1}^N p(x_n|\mu)=\prod_{n=1}^N \mu^{x_n}(1-\mu)^{1-x_n}$

**conjugacy**

if we choose a prior $p(\mu)$ has the form of $[\text{something} \cdot \mu^x(1-\mu)^{1-x}]$

our posterior $p(\mu|\mathcal{D})$ can also have a form of $[\text{something} \cdot \mu^x(1-\mu)^{1-x}]$

suppose a Binomial likelihood function, Binomial is a generalization of Bernoulli:

$Bin(m|N,\mu)=C_N^m \mu^m (1-\mu)^{N-m} \propto \mu^m (1-\mu)^{l}$, where $l=N-m$

$
\begin{align}
p(\mu|m,l,a,b) &\propto Bin(m|l,\mu) \times Beta(\mu|a,b) \\
&\propto \mu^m (1-\mu)^{l} \times \mu^{a-1}(1-\mu)^{b-1} \\
&\propto \mu^{m+a-1} (1-\mu)^{l+b-1}  \\
\end{align}
$

$p(\mu|m,l,a,b)$ is simply another **Beta distribution**:

$p(\mu|m,l,a,b)=\frac{\Gamma(m+a+l+b)}{\Gamma(m+a)\Gamma(l+b)}\mu^{m+a-1} (1-\mu)^{l+b-1}$

recall Beta distribution:

$Beta(\mu|a,b)=\frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)}\mu^{a-1}(1-\mu)^{b-1}$

$a,b$ can be interpreted as an **effective number of observations of x=1 and x=0**

**further more, the posterior distribution can act as the prior if we subsequently observe additional data**

for example, in one-step Bayesian, we have 

prior: $Beta(\mu|a=2,b=2)$

likelihood: $Bin(m=1|N=1,\mu)=C_1^1\mu$, corresponds to a single observation of x=1

posterior: $Beta(\mu|a=3,b=2)$

### curve fitting example

for $N$ observations/data points

$\textbf{x} \triangleq (x_1,...,x_N)^T$

$\textbf{t} \triangleq (t_1,...,t_N)^T$

**Goal**: for a new $\hat{x}$, predict $\hat{t}$

use **polynomial function** for $M$ features/order of the polynomial

$h(\textbf{w})=w_0+w_1x^2+w_2x^3...+w_M x^M=\sum_{j=0}^{M} w_jx^j$

where $\textbf{w}=(w_0,...,w_M)^T$

**Minimize the Cost/Error functions**:

$J(\textbf{w})=\frac{1}{2}\sum_{n=1}^N [h_n(\textbf{w})-t_n]^2$

for regularization

$J(\textbf{w})=\frac{1}{2}\sum_{n=1}^N [h_n(\textbf{w})-t_n]^2+\frac{\lambda}{2}||\textbf{w}||^2$

where $||\textbf{w}||^2=\textbf{w}^T\textbf{w}$

so far the **model complexity** is controlled by $M$ and $\lambda$

**Bayesian View**:

model the uncertainty of target $\textbf{t}$ with Gaussian

for each data:

$p(t|x,\textbf{w},\beta)=\mathcal{N}(t| h(\textbf{w}),\beta^{-1})$

for the training dataset, the **likelihood** function:

$p(\textbf{t}|\textbf{x},\textbf{w},\beta)=\prod_{n=1}^N \mathcal{N}(t_n| h_n(\textbf{w}),\beta^{-1})$

let's max the log of likelihood:

$\log p(\textbf{t}|\textbf{x},\textbf{w},\beta)= -\frac{\beta}{2}\sum_{n=1}^N [h_n(\textbf{w})-t_n]^2+\frac{N}{2}\log \beta - \frac{N}{2} \log (2\pi)$

to obatin $\textbf{w}_{ML}$ and $\beta_{ML}$

now we can make predictions for new values of $x$ by the **predictive distribution**

$p(t|x,\textbf{w}_{ML},\beta_{ML})=\mathcal{N}(t|h(\textbf{w}_{ML}),\beta_{ML}^{-1})$

**Further**:

we can introduce a **prior** $p(\textbf{w}|\alpha)=\mathcal{N}(\textbf{w}|\textbf{0},\alpha^{-1}\textbf{I})=\left(\frac{\alpha}{2\pi} \right)^{(M+1)/2}\exp\left[-\frac{\alpha}{2} \textbf{w}^T\textbf{w} \right]$

$\alpha$ is another Gaussian precision similar as $\beta$, by Baye's Theorem

$p(\textbf{w}|\textbf{x},\textbf{t},\alpha,\beta) \propto p(\textbf{t}|\textbf{x},\textbf{w},\beta)p(\textbf{w}|\alpha)$

now we can use **maximum posterior (MAP)** to determine $\textbf{w}$ by finding the most probable value of $\textbf{w}$ given data

to $\max p(\textbf{w}|\textbf{x},\textbf{t},\alpha,\beta)$ we can take the negative log of $p(\textbf{t}|\textbf{x},\textbf{w},\beta)p(\textbf{w}|\alpha)$, which gives us 

$\min \frac{\beta}{2} \sum_{n=1}^N [h(\textbf{w})-t_n]^2+\frac{\alpha}{2} \textbf{w}^T\textbf{w}$

**a full Bayesian approach**:

we wish to evaluate the prediction distribution $p(t|x,\textbf{x},\textbf{t})$

assume that $\alpha$ and $\beta$ are fixed and known

based on the Bayesian laws

$p(t|x,\textbf{x},\textbf{t})=\int p(t|x,\textbf{w})p(\textbf{w}|\textbf{x},\textbf{t})d\textbf{w}$

where

$p(t|x,\textbf{w})$ is given by $p(t|x,\textbf{w},\beta)=\mathcal{N}(t| h(\textbf{w}),\beta^{-1})$

$p(\textbf{w}|\textbf{x},\textbf{t})$ is given by normalized $p(\textbf{w}|\textbf{x},\textbf{t},\alpha,\beta) \propto p(\textbf{t}|\textbf{x},\textbf{w},\beta)p(\textbf{w}|\alpha)$

we can obtain an analytical result of $p(t|x,\textbf{x},\textbf{t})$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(2)
x=np.linspace(0,1,10) #sample size=10

#std 0.25
t=np.sin(2*np.pi*x)+np.random.normal(scale=0.25,size=x.shape)

x_train,y_train=x,t

x_true=np.linspace(0,1,100)
y_true=np.sin(2*np.pi*x_true)

plt.scatter(x_train,y_train,facecolor='none',edgecolor='b',s=50,label='training data')
plt.plot(x_true,y_true,c='g',label='$\sin(2\pi x)$')
plt.legend()
```

### the cancer X-Ray classification example

$\textbf{x}$ - the set of pixel intensities in the image

$\textbf{t}$ - cancer or not

$t \in C_1$ - cancer

$t \in C_2$ - no cancer

the **inference step** involves determining the joint distribution $p(\textbf{x},C_k)$

the **decision step** has to determine whether to give a treatment

by Baye's Theorem

$p(C_k|\textbf{x})=\frac{p(\textbf{x}|C_k)p(C_k)}{p(\textbf{x})}$

$p(\textbf{x},C_k)$ is important to obtain quantities in the above items

i.e. $p(\textbf{x},C_k)=p(C_k|\textbf{x})p(\textbf{x})=p(\textbf{x}|C_k)p(C_k)$

$p(C_1)$ - the probability that a person has cancer before X-ray measurement

$p(C_1|\textbf{x})$ - the probability that a person has cancer when we have X-ray info

**goal** is to minimize the chance of assigning $x$ to the wrong class, we can do this in several ways

- minimizing the misclassification rate

$
\begin{align}
\min p(mistake) &=p(\textbf{x} \in R_1, C_2)+p(\textbf{x} \in R_2, C_1)\\
& = \int_{R_1} p(\textbf{x},c_2) d\textbf{x} + \int_{R_2} p(\textbf{x},c_1) d\textbf{x} \\
& = p_{R_1} (C_2) + p_{R_2} (C_1)
\end{align}
$

to minimize $p(mistake)$, we should arrange $x$ to be assigned to whichever class has the smaller value of $p(\textbf{x},c_2)$ and $p(\textbf{x},c_1)$, thus if $p(\textbf{x},c_1)>p(\textbf{x},c_2)$ for a given $x$, we should assign that $x$ to $C_1$

since $p(\textbf{x},c_k)=p(c_k|\textbf{x})p(\textbf{x})$, $p(\textbf{x})$ is same for each $k$,

min the mistake rate <=> max $p(c_k|\textbf{x})$

- maximizing the correct rate

$
\begin{align}
\max p(correct) &= \sum_{k=1}^K p(\textbf{x} \in R_k, C_k)\\
& = \sum_{k=1}^K \int_{R_k} p(\textbf{x}, C_k) d\textbf{x} \\
& = p_{R_1} (C_1) + p_{R_2} (C_2) + ...
\end{align}
$

- minimizing the expected loss

suppose the true label $C_k$, and the predicted label $C_j$ for a given $x$

we can denote the loss by $L_{kj}$

|   |j=1 |j=2 |
|---|----|----|
|k=1| 0  |1000|   
|k=2| 1  | 0  |

the optimal solution is to minimize the loss function $L_{kj}$, but we don't know the true class $C_k$, since we are given a new data $x$ to predict $C_k$

we instead minimize the average loss by choosing the regions $R_j$:

$\min \mathbb{E}[L]=\sum_k \sum_j \int_{R_j} L_{kj}p(\textbf{x},C_k) d\textbf{x}$

minimizing this loss is equivalent to minnimize 

$\min \sum_k L_{kj}p(C_k|\textbf{x})$



### MLE example of Kumaraswamy distribution

https://python.plainenglish.io/scipy-has-failed-you-do-maximum-likelihood-estimation-by-yourself-34145d71b63f

pdf:

$f(x)=abx^{a-1}(1-x^a)^{b-1}$

let's say we have a data set $\textbf{x}$ with $n$ points was suspected drawn from the Kumaraswamy distribution

if we assume our data is i.i.d

we can write the likelihood function:

$
\begin{align}
L(a,b|\textbf{x})&=[abx_1^{a-1}(1-x_1^a)^{b-1}][abx_2^{a-1}(1-x_2^a)^{b-1}]...[abx_n^{a-1}(1-x_n^a)^{b-1}] \\
& = (ab)^n \prod_{i=1}^n x_i^{a-1}(1-x_i^a)^{b-1}
\end{align}
$

then we take log of the likelihood

$\log L(a,b|\textbf{x})=N\log(ab)+(a-1)\sum_{i=1}^N \log(x_i)+(b-1)\sum_{i=1}^N \log(1-x_i^a)$

**sometimes, the log-likelihood function has nice closed-form solutions for the parameters, but sometimes not, in most cases, we obtain complicated forms which must be solved numerically**

the Kumaraswamy quantile function for generating samples:

$Q(p,a,b)=(1-(1-p)^{1/b})^{1/a}$


```python
import scipy.optimize as opt
import scipy.stats as st
import numpy as np
import seaborn as sns

#use log of parameters to avoid the potential erros due to
#invalid values placed in the log of kumaraswamy_ll function 
#during the optimization 
def kumaraswamy_ll(log_params,x):
    
    #ll short for loglikelihood
    #a,b, are parameters
    N=len(x)
    a,b=np.exp(log_params)
    ll=N*np.log(a*b)+(a-1)*np.sum(np.log(x))+(b-1)*np.sum(np.log(1-np.power(x,a)))
    
    return ll

def kumaraswamy_mle(x):
    
    res=opt.minimize(fun=lambda log_params, x: -kumaraswamy_ll(log_params,x),
                     x0=np.array([0.5,0.5]),
                     args=(x,),
                     method='BFGS')
    
    a,b=np.exp(res.x)
    
    return a,b
    
def kumaraswamy_q(p,a=10,b=2):
    
    return np.power(1-np.power(1-p,1/b),1/a)

#sample from uniform distribution
sample=st.uniform.rvs(0,1,20000)
kumaraswamy_sample=kumaraswamy_q(sample,a=10,b=2)

sns.histplot(kumaraswamy_sample)
```

```python
a,b=kumaraswamy_mle(kumaraswamy_sample)
sns.histplot(kumaraswamy_sample)
x=np.linspace(0,1,num=1000)
plt.plot(x,170*a*b*x**(a-1)*(1-x**a)**(b-1),'r',linewidth=3)
```

### locally weighted linear regression

$\sum_i^n w^{(i)} [y^{(i)}-\theta^T x^{(i)}]^2$

$n$ - number of training data

$w^{(i)}=\exp{\left[-\frac{(x^{(i)}-x)^T (x^{(i)}-x)}{2\tau^2} \right]}$

a closed form solution

$\theta=(X^TWX)^{-1}(X^TWY)$

ref:

https://towardsdatascience.com/locally-weighted-linear-regression-in-python-3d324108efbf


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('default')

np.random.seed(8)
X=np.random.randn(1000,1)
y=2*(X**3) + 10 + 4.6*np.random.randn(1000,1)

plt.plot(X,y,'o',markeredgecolor='w')

def weight_matrix(pt,X,tau):
    
    m=X.shape[0]
    w=np.mat(np.eye(m))
    
    for i in range(m):
        xi=X[i]
        d=(-2*tau*tau)
        w[i,i]=np.exp(np.dot((xi-pt),(xi-pt).T)/d)
        
    return w

def predict(X,y,pt,tau=0.08):
    
    m=X.shape[0]
    
    #adding bias term
    X_=np.append(X,np.ones(m).reshape(m,1),axis=1)
    
    pt_=np.array([pt,1])
    
    w=weight_matrix(pt_,X_,tau)
    
    theta=np.linalg.pinv(X_.T*(w*X_))*(X_.T*(w*y))
    
    pred=np.dot(pt_,theta)
    
    return theta,pred

```

```python
n=300
X_test=np.linspace(-3,3,n)

preds=[]
for pt in X_test:
    
    theta,pred=predict(X,y,pt,tau=0.1)
    preds.append(pred)
    
X_test=np.array(X_test).reshape(n,1)
preds=np.array(preds).reshape(n,1)

plt.plot(X,y,'o',markeredgecolor='w')
plt.plot(X_test,preds,'ro')
```

```python
def sigmoid(x):
    
    #can be applied to all eles in a np array because of np.exp
    return 1/(1+np.exp(-x))

def cost(X,y,w):
    
    n=len(y)
    h=sigmoid(X@w)
    eps=1e-5
    
    
    #+eps prevent log underflow
    #-logloss
    return ((1/n)*(((-y).T@np.log(h+eps))-((1-y).T@np.log(1-h+eps))))[0][0]

def gradient_descent(X,y,w,lr,n_iters):
    
    n=len(y)
    costs=[]
    
    for i in range(n_iters):
        
        w=w-(lr/n)*(X.T@(sigmoid(X@w)-y))
        costs.append(cost(X,y,w))
        
    return costs,w

def predict(X,w):
    
    #round up probabilities
    return np.round(sigmoid(X@w))

from sklearn.datasets import make_classification
X,y=make_classification(n_samples=500, n_features=2, n_redundant=0, n_informative=1,
                        n_clusters_per_class=1, random_state=14)

```
